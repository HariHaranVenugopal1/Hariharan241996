```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df_1 = pd.read_csv(r"C:\Users\hp\OneDrive\Desktop\Primary.csv")
df_1

remove_df_1 = df_1.drop(0)
remove_df_1

remove_df_1.columns = ['ISO3','Countries and areas','Region','Sub-region','Income Group','Total(%)','Rural Residence(%)', 'Urban Residence(%)','Poorest Wealth quintile(%)','Richest Wealth quintile(%)', 'Data source','Time period']
remove_df_1

drop_df_1 = remove_df_1.drop(columns = ['Data source'])
drop_df_1

drop_df_1['Total(%)'] = drop_df_1['Total(%)'].str.replace('%','')
print(drop_df_1)

drop_df_1['Total(%)'] = drop_df_1['Total(%)'].astype(int)
drop_df_1

drop_df_1['Rural Residence(%)'] = drop_df_1['Rural Residence(%)'].str.replace('%','')
print(drop_df_1)

drop_df_1['Urban Residence(%)'] = drop_df_1['Urban Residence(%)'].str.replace('%','')
print(drop_df_1)

drop_df_1['Poorest Wealth quintile(%)'] = drop_df_1['Poorest Wealth quintile(%)'].str.replace('%','')
print(drop_df_1)

drop_df_1['Richest Wealth quintile(%)'] = drop_df_1['Richest Wealth quintile(%)'].str.replace('%','')
print(drop_df_1)

fill_na_1 = drop_df_1.fillna(0)
print(fill_na_1)

remove_df_1 = fill_na_1.isna().sum().reset_index()
print(remove_df_1)

fill_na_1['Time period'] = fill_na_1['Time period'].apply(lambda x: x.split('-')[0])

fill_na_1['Time period'] = fill_na_1['Time period'].astype(int)
remove_df_1

fill_na_1 =fill_na_1[fill_na_1['Time period'] <= 2022]
fill_na_1

fill_na_1['Countries and areas'] = fill_na_1['Countries and areas'].replace({'C™te d\'Ivoire': 'cote d\'ivoire'}, regex=True)
fill_na_1

cleaned_Primary_csv = fill_na_1.to_csv('cleaned Primary.csv', index = True)
print('\nCSV String:\n', cleaned_Primary_csv)
```
